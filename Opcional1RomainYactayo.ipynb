1. Histograma de Ataque y de Defensa en 2 subgráficos
Se deberán emplear las variables “attack” y “defense” y la función de generación de histogramas. Se deberá representar cada uno en un subgráfico diferente.

```
#SubGráficas
import matplotlib.pyplot as plt1
fig, axarr = plt1.subplots(2,1, figsize=(12,8))

poke[poke['attack']<200]['attack'].plot.hist(ax=axarr[0])
poke[poke['defense']<400]['defense'].plot.hist(ax=axarr[1])
```

```
#SubGráficas 2
import matplotlib.pyplot as plt1
fig, axarr = plt1.subplots(2,2, figsize=(12,10))

poke1 = poke.groupby(['type1']).mean()
poke2=poke1['base_happiness']
poke3=poke1['percentage_male']
poke4=poke1['weight_kg']
poke5= poke1[['base_happiness','percentage_male']]



poke2.plot.bar(stacked=False,ax=axarr[0][0])
poke3.plot.bar(stacked=False,ax=axarr[0][1])
poke4.plot.bar(stacked=False,ax=axarr[1][0])
poke5.plot.hexbin(x='base_happiness',y='percentage_male', gridsize=15, ax=axarr[1][1])
```
